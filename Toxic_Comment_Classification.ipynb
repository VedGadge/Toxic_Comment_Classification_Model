```python
MAX_FEATURES = 200000 # number of word in the vocabulray
```

```python
## output_sequence_length --> Max length of our sentence, sentence length in token
## Each sentence from our input,  length will be capped to 1800 words
## output_mode --> Map every single word to an integer
```

```python
vectorizer = TextVectorization(max_tokens = MAX_FEATURES, output_sequence_length= 1800, output_mode='int')
```

### Note: -
The vocabulary for the layer must be either supplied on construction or learned via adapt(). When this layer is adapted, it will analyze the dataset, determine the frequency of individual string values, and create a vocabulary from them. This vocabulary can have unlimited size or be capped, depending on the configuration options for this layer; if there are more unique values in the input than the maximum vocabulary size, the most frequent terms will be used to create the vocabulary.

```python
#Tensorflow datapipeline
# MCSBAP - MAP, CHACHE, SHUFFLE, BATCH, PREFETCH from tensor_slices, list_files
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) # helps prevet bottlenecks
```

```python
train = dataset.take(int(len(dataset)*0.7)) # take 70% of dataset for training
val = dataset.skip(int(len(dataset)*0.7)).take(int(len(dataset)*0.2)) # skip the 70% datset and then take 20% from other part for validation set
test = dataset.skip(int(len(dataset)*0.9)).take(int(len(dataset)*0.1)) # skip 90% and take the remaining 10% for training test
```

## Create the Sequential Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
model = Sequential()
# create the embedding layer
model.add(Embedding(MAX_FEATURES + 1, 32)) # no of words +1(for unknown words), 32 length of each embedding
# Create Bidirectional LSTM layer
model.add(Bidirectional(LSTM(32, activation = 'tanh'))) # Bidirectional LSTM is needed particulary for sentences(may need previous modifier words)
# Feature extractor Fully Connected Layer
model.add(Dense(128, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
#Final layer --> We have 6 different outputs
model.add(Dense(6, activation = 'sigmoid'))
```

```python
y[0]
```

```python
model.compile(loss = 'BinaryCrossentropy', optimizer = 'Adam')
# No categorical crossentropy ? 6outputs --> no ,its not like we will get 1 value from the 6 classes
# It is like we are running 6 different binary classifiers at one time
```

```python
model.summary()
```

```python
history = model.fit(train, epochs = 1, validation_data = val)
```

## Evaluate Model

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    # Unpack the batch 
    X_true, y_true = batch
    # Make a prediction 
    yhat = model.predict(X_true)
    
    # Flatten the predictions
    y_true = y_true.flatten()
    yhat = yhat.flatten()
    
    pre.update_state(y_true, yhat)
    re.update_state(y_true, yhat)
    acc.update_state(y_true, yhat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```

```python
def score_comment(comment):
    vectorized_comment = vectorizer([comment])
    results = model.predict(vectorized_comment)
    
    text = ''
    for idx, col in enumerate(df.columns[2:]):
        text += '{}: {}\n'.format(col, results[0][idx]>0.5)
    
    return text
```

```python
interface = gr.Interface(fn=score_comment, 
                         inputs=gr.inputs.Textbox(lines=2, placeholder='Comment to score'),
                        outputs='text')
```
